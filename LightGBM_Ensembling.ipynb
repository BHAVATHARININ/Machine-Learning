```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Breast_cancer_data.csv")
```

```python
x = df.drop(columns = ['diagnosis'])
y = df['diagnosis']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
pip install lightgbm
```

```python
import lightgbm as lgb
```

```python
model = lgb.LGBMClassifier(n_estimators = 500, max_depth = 2, learning_rate = 1.0)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred_train = model.predict(x_train)
acc_train = accuracy_score(y_train, y_pred_train)
print("Training Accuracy : ",acc_train)
```

```python
y_pred_test = model.predict(x_test)
acc_test = accuracy_score(y_test, y_pred_test)
print("Testing Accuracy : ", acc_test)
```

```python
print("Classification Report\n")
print(classification_report(y_test, y_pred_test))
```

Same Model using Grid Search Approach for HyperParameter tuning

```python
import lightgbm as lgb
model = lgb.LGBMClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'learning_rate' : [0.25,0.50,0.75,1.0],
    'max_depth' : [2,3,4,5],
    'n_estimators' : [200,300,400,500],
    'num_leaves' : [20,30,40]

}
grid_search = GridSearchCV(estimator = model, param_grid = param_grid, scoring = 'accuracy', cv = 5, verbose=1)
```

```python
grid_search.fit(x_train, y_train)

print("Best Parameters : ",grid_search.best_params_)
print("Best Accuracy : ", grid_search.best_score_)
```
