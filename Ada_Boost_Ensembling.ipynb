```python
columns_to_drop = ['Name','Age', 'SibSp', 'Ticket', 'Parch', 'Cabin', 'Embarked']
train_test.drop(labels = columns_to_drop, axis = 1, inplace = True)
```

```python
train_test_dummies = pd.get_dummies(train_test, columns=['Sex'])
```

```python
train_test_dummies
```

```python
train_test_dummies.isnull().sum()
```

```python
train_test.fillna(value = 0.0, inplace = True)
```

```python
train_test.isna().sum
```

```python
x_train = train_test_dummies.values[0:891]
x_test = train_test_dummies.values[891:]
```

```python
train_test_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
```

```python
x_train_scale
```

```python
x_test_scale
```

```python
from sklearn.model_selection import train_test_split
x_train_sub, x_validation_sub, y_train_sub, y_validation_sub = train_test_split(x_train_scale, y_train, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
ada = AdaBoostClassifier(n_estimators = 20,random_state = 0)
ada.fit(x_train_sub, y_train_sub)
print("Accuracy Score (training) : {0:.3f}".format(ada.score(x_train_sub, y_train_sub)))
print("Accuracy score (validation) : {0:.3f}".format(ada.score(x_validation_sub, y_validation_sub)))
```
